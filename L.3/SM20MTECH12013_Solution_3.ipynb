### Part a) Since all the fuctions involved here are linear and as we know that the linear functions are convex also, so the minimum initial investment that funds the expense stream can be found by solving the following convex optimization problem.

```
#Define_Variable
b = cp.Variable((n,1))
w = cp.Variable((n,1))
x = cp.Variable((m,1))

#Define_problem

#objective
z = b[1]+(np.ones(m)).T@x
objective = cp.Minimize(z)

#contraints
constr = [b>=0 , w+P@x>=e, x>=0 ]
constr += [b[t+1]==(1+rho)*b[t]-w[t] for t in range (n-1)]

prob = cp.Problem(objective,constr)
prob.solve()
print("Minimum initial investment in dollars:" ,z.value)
```

```
constr += [x==0]
prob = cp.Problem(objective,constr)
prob.solve()
print("Minimum initial investment in dollars when x=0:" ,z.value)
```
